```python
import pandas as pd
```

```python
df = pd.read_csv("fraudTrain.csv")
# df = pd.read_csv("fraudTest.csv")
df.head()
```

```python
total1 = len(df)
print("总数据量（包括重复的）：", total1)
```

```python
df.drop_duplicates(inplace=True)
```

```python
total2 = len(df)
print("总数据量（包括重复的）：", total2)
```

```python
df.drop(["Unnamed: 0", "cc_num", "merchant", "gender", "street", "city","state","zip", "job", "trans_num", "unix_time"], axis=1, inplace=True)
df.head()
```

```python
# 查看 "first" 和 "last" 列都相同的唯一值数量
unique_count = df.groupby(['first', 'last']).ngroups
print("'first' 和 'last' 列都相同的唯一值数量：", unique_count)

# 输出总数据量（包括重复的）
total_count = len(df)
print("总数据量（包括重复的）：", total_count)
```

```python
# 查看 category 列的唯一值及其数量
value_counts = df['category'].value_counts()
print("category 列的唯一值及其数量：")
print(value_counts)
```

```python
# 过滤出 is_fraud=0 的行
filtered_df = df[df['is_fraud'] == 0]

# 按 category 分组，并计算 amt 的平均值
avg_amt_by_category = filtered_df.groupby('category')['amt'].mean()
avg_amt_by_category.head()
```

```python
# 按 category 分组，并计算 amt 的中位数
median_amt_by_category = filtered_df.groupby('category')['amt'].median()
median_amt_by_category.head()
```

```python
# 将中位数映射到原始 DataFrame 中
# df['category_median'] = df['category'].map(median_amt_by_category)

# 计算 amt 与中位数的差值
df['amt_diff'] = df['amt'] - df['category'].map(median_amt_by_category)
df.head()
```

```python
# 将日期列转换为 datetime 类型
df['dob'] = pd.to_datetime(df['dob'])

# 提取出生年份
df['dob'] = df['dob'].dt.year

# 计算年龄
current_year = 2025
df['dob'] = current_year - df['dob']
```

```python
# 将日期时间列转换为 datetime 类型
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

# 提取日期、时间、月份和星期几
df['time'] = df['trans_date_trans_time'].dt.hour
df['month'] = df['trans_date_trans_time'].dt.month
df['weekday'] = df['trans_date_trans_time'].dt.weekday  # 获取星期几的数字表示（0=星期一, 6=星期日）
```

```python
df['Llat'] = abs(df['merch_lat'] - df['lat'])
df['Llong'] = abs(df['merch_long'] - df['long'])

# 删除原始的 'lat', 'long', 'merch_lat', 'merch_long' 列
df.drop(['lat', 'long', 'merch_lat', 'merch_long','trans_date_trans_time'], axis=1, inplace=True)

df.head()
```

```python
# # 计算每个类别的目标均值
# target_encoding = df.groupby('category')['is_fraud'].mean()

# # 将目标均值映射到原始数据
# df['Encoding'] = df['category'].map(target_encoding)
import pandas as pd
# One-Hot 编码
df = pd.get_dummies(df, columns=['category'], drop_first=True)
```
